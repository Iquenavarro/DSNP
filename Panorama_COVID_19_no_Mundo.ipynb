# Panorama da COVID-19 no mundo e a influência da vacinação no controle da doença

  A COVID-19 é uma doença infectocongatiosa, provocada pelo coronavírus da síndrome respiratória aguda grave 2 (SARS-CoV-2). Conforme declaração da Organização Mundial da Saúde (OMS) no dia 31 de Dezembro de 2019, na cidade de Wuhan na China, foram identificados os primeiros casos de uma pneumonia de origem desconhecida. 
  
  No dia 07 de Janeiro de 2020 o virus estava sequenciado pela as autoridades de saúde e disponíveis no banco de dados da OMS. Logo após, no dia 13 de Janeiro a Tailândia resgistra o seu primeiro caso, seguida do Japão (15 de Janeiro) e Coreia do Sul (20 de Janeiro) chegando nos Estados Unidos da América (EUA) no dia 23 de Janeiro.

  Em 2019, não foi a primeira ocorrência do vírus, sendo a primeira vez em 1937 em que ele foi isolado e ficou conhecido em 2002, a primeira vez em que causou uma síndrome respiratória aguda grave (SARS) causando uma insuficiência respiratória e rápidamente contidada por EUA, Canadá e China.

  O SARS-CoV-2 é menos letal que algumas outras variantes que circulam no Oriente Médio (MERS-CoV) com Origem na Árabia Saudita em 2012. Porém, o que faz o SARS-CoV-2 é o potencial de contaminação.

  A transmissão do Vírus ocorre por gotículas, secreções respiratórias e contatos com pessoas infectadas. Um estudo de Van Doremalen et al. identificou que o SARS-CoV-2 pode permanecer ativo e com capacidade de infectar por até 3h após eliminado do local, a depender da quantidade de virus secretado e da superfície de deposição. A transmissão direta ainda permanece como o principal meio de contágio, mas a indireta - via superfícies infectadas - também tem sua parcela de contribuição. 

  Como forma de contribuição para a ciência e a consciencia populacional acerca dos riscos, será realizada uma análise de dados dos casos da doença, bem como do avanço da vacinação.

```python
# histograma dos casos

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)

x = df.loc[df['location']== 'World','date']
y = df.loc[df['location']== 'World', 'total_cases']

myFmt = mdates.DateFormatter("%Y-%m")

for ax in (ax1, ax2):
    ax.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
    ax.xaxis.set_major_formatter(myFmt)
    ax.tick_params(axis='x', rotation=70)
    ax.set_ylabel('Casos')
    ax.fill_between(x, y, alpha=0.25)
    
ax1.plot(x, y)
ax1.set_title('Escala linear')
ax1.yaxis.set_major_formatter(EngFormatter())

ax2.plot(x, y)
ax2.set_yscale('log')
ax2.yaxis.set_major_formatter(EngFormatter())
ax2.set_title('Escala logarítmica')


plt.suptitle('Evolução do total de casos de COVID-19 no mundo\n', 
             color='dimgray')


plt.show()
```

## E o Brasil? 

```python
variants = {
    'alfa': '2020-09-01',
    'beta': '2020-05-01',
    'gama': '2020-11-01',
    'delta': '2020-10-01',
    'lambda': '2020-12-01',
    'mu': '2021-01-01',
}

variants = {key: datetime.fromisoformat(value) for (key, value) in variants.items()}

colors = plt.cm.Dark2.colors
```

```python
fig, ax = plt.subplots(figsize=(15,10))

x = df.loc[df['location'] == 'Brazil', 'date']
y2 = df.loc[df['location'] == 'Brazil', 'new_deaths_smoothed']

ax.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
ax.xaxis.set_major_formatter(myFmt)
ax.tick_params(axis='x', rotation=70)
ax.set_ylabel('Novos casos')


ax.plot(x, y2)
ax.yaxis.set_major_formatter(EngFormatter())
ax.axvspan(*mdates.datestr2num(['2020-06-01', '2020-08-31']),color='yellow', alpha=0.25, label = 'inverno')
ax.axvspan(*mdates.datestr2num(['2021-06-01', '2021-08-31']), color='yellow', alpha=0.25)

for color, var in zip(colors, variants):
    ax.axvline(x=variants[var], label=var, linestyle='--', color=color)

plt.suptitle('Evolução de novas mortes no Brasil', color='dimgray', fontsize=22)

#plt.gcf().text(0.4, 0.885, 'Highlights nas datas do inverno e de variantes')

handles, labels = ax.get_legend_handles_labels()

fig.legend(bbox_to_anchor=(0.4,0.885), loc='upper center', ncol=8, handles=handles, labels=labels, fontsize=11)

plt.show()

```
